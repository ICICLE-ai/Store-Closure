```python
import pandas as pd
import re

stores = pd.DataFrame(columns=["id","name","type","latitude","longitude"])

file_name_list = [
    "data/CurbPickup541110Franklin.csv",
    "data/EthnicFoods541107Franklin.csv",
    "data/GrocerRetail541105Franklin.csv",
    "data/HealthFoods541108Franklin.csv",
    "data/ShoppingService729941Franklin.csv",
    "data/SpecialtyFoods549910Franklin.csv",
    "data/SpecialtyFoods549918Franklin.csv",
    "data/Wholesale514105Franklin.csv"
]

count = 0
for file in file_name_list:
    items = pd.read_csv(file)
    for index,row in items.iterrows():
        pattern = "data/([A-Za-z]+)"
        match = re.search(pattern, file).group(1)
        print(match)
        stores.loc[count] = {"id":count,"name":row["Company Name"],"type":match,"latitude":row["Latitude"],"longitude":row["Longitude"]}
        count+=1

stores.to_csv("data/stores.csv",index=False)
print(stores)
```
